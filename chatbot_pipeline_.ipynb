```python
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration
from huggingface_hub import login

# Authenticate (Only needed once)
login(token="hf_haWwszqGufeMxQw")  # ⚠️ Replace with your token

# Load model and tokenizer
model_name = "facebook/blenderbot-3B"
tokenizer = BlenderbotTokenizer.from_pretrained(model_name, use_auth_token=True)
model = BlenderbotForConditionalGeneration.from_pretrained(model_name, use_auth_token=True)

```

```python
def chat():
    print("🤖 Chatbot: Hello! Type 'exit' to end the conversation.")
    conversation_history = ""

    while True:
        user_input = input("🧑 You: ")
        if user_input.lower() == "exit":
            print("🤖 Chatbot: Goodbye! Have a great day!")
            break

        # Append the user input to conversation history
        conversation_history += " " + user_input

        # Generate response
        inputs = tokenizer(conversation_history, return_tensors="pt")
        reply_ids = model.generate(**inputs)
        bot_reply = tokenizer.decode(reply_ids[0], skip_special_tokens=True)

        # Display bot response and update conversation history
        print(f"🤖 Chatbot: {bot_reply}")
        conversation_history += " " + bot_reply  # Keep the flow of conversation

# Start chatting
chat()
```
